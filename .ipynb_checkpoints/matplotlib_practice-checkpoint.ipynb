```python
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plt() a object oriented method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,marker="*", color="green", linewidth=2, label='Boston')
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.grid()
```

```python
#create the bar plot
plt.bar(x_axis, y_axis)
```

```python
plt.barh(x_axis,y_axis,color="magenta",label='Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Fare')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
```

```python

#create a plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.barh(x_axis,y_axis,color="cyan",label='Chicago')
ax.legend()
ax.set_xlabel("Fare")
ax.set_ylabel("Month")
ax.invert_yaxis()
```

```python
plt.scatter(y_axis,x_axis,c="red",label='Chicago',marker ='x',s=y_axis)

plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,50)
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.show()
```

```python
y_axis_large =[]
for data in y_axis:
    y_axis_large.append(data*3)
```

```python
plt.scatter(y_axis,x_axis,c="red",label='Chicago',marker ='x',s=y_axis_large)

plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,50)
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.show()
```
