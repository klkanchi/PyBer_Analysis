```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plt() a object oriented method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,marker="*", color="green", linewidth=2, label='Boston')
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.grid()
```

```python
#create the bar plot
plt.bar(x_axis, y_axis)
```

```python
plt.barh(x_axis,y_axis,color="magenta",label='Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Fare')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
```

```python

#create a plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.barh(x_axis,y_axis,color="cyan",label='Chicago')
ax.legend()
ax.set_xlabel("Fare")
ax.set_ylabel("Month")
ax.invert_yaxis()
```

```python
plt.scatter(y_axis,x_axis,c="red",label='Chicago',marker ='x',s=y_axis)

plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,50)
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.show()
```

```python
y_axis_large =[]
for data in y_axis:
    y_axis_large.append(data*3)
```

```python
plt.scatter(y_axis,x_axis,c="red",label='Chicago',marker ='x',s=y_axis_large)

plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0,50)
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.show()
```

```python
#using list comprehension inside scatter so that for loop is inside the scatter
plt.scatter(x_axis,y_axis,s=[i * 3 for i in y_axis])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(y_axis,x_axis,color="skyblue",s=[i*5 for i in y_axis],alpha=0.2,edgecolors="black",linewidths=2)
ax.invert_yaxis()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 10))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0.6, 0, 0, 0.2, 0, 0, 0, 0, 0)


ax.pie(y_axis,
      explode = explode_values,
      colors = colors,
      labels = x_axis,
        autopct = '%.1f%%',
       shadow = True
      )
```

```python
import statistics
mean = statistics.mean(y_axis)
mean
stdev= statistics.stdev(y_axis)
stdev
```

```python
ax.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
fig,ax = plt.subplots()

plt.show()
```

```python
import numpy as np
fig,ax = plt.subplots()
ax.bar(x_axis,y_axis)
ax.set_yticks(np.arange(0,51,step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
